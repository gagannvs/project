```python
import numpy as np
import pandas as pd

data=pd.read_csv('/home/chaitra/Desktop/finds.csv')
data
```

```python
def train(concepts,target):
    #specific_h=concepts[0]#assigning first concept as specific hypothesis
    if target[0]=='yes':
        specific_h=concepts[0]
    else:
        specific_h=[0,0,0,0,0,0]
    print('specific1',specific_h)
    for i,h in enumerate(concepts):#enumerate function: gives index and value. Here 'i' is index and 'h' is hypothesis
        if target[i]=="Yes":
            for x in range(len(specific_h)):#for all columns--checking for one-one value like sunny
                if h[x]==specific_h[x]:
                    pass
                elif specific_h[x]==0:
                    specific_h[x]=h[x]
                else:
                    specific_h[x]="?"
    return specific_h
concepts=np.array(data.iloc[:,0:-1])#data slicing: Except last(Target column),slicing col
target=np.array(data.iloc[:,-1])#slicing of last column 
print('con', concepts)
print('Tar',target)
print(train(concepts,target))
```
